
# Назначение проектируемой системы

Система агрегации и бронирования ресторанов. Данная система предназначена для быстрого нахождения и бронирования 
ресторана, который интересует клиента.      

# Функциональные требования

## Участники

+ Клиент (Client):
 пользователь системы для бронирования ресторанов.
  - регистрация
  - поиск ресторана по имени
  - поиск ресторана по кухне
  - формирование заявки о бронировании
  - отказ от бронирования
  - отзыв о посещении

+ Представитель (Representative):
 работник ресторана, который рассматривает заявки о бронировании, меняет описание ресторана. 
  - регистрация
  - формирование карточки ресторана
  - анализ поступающих заявок на бронь
  - ответ на заявку по брони
  - подтверждение посещения клиента

+ Администратор (Admin):  
 работник системы, который модерирует информацию в системе и разрешает споры между клиентом и рестораном.  
  - регистрация
  - регистрация нового ресторана
  - разрешение споров по отзывам


## Описание бизнес-процессов

### 1. Регистрация нового ресторана (Restaurant's Sign up):

#### Участники:
- администратор
- представитель

#### Этапы:
- создание записи ресторана
- отправление сопутствующего письма с цифровой подписью ресторана на почту системы
- рассмотрение заявки администратором системы
- запрос дополнительных документов администратором
- подтверждение регистрации ресторана
- создание учетной записи представителя ресторана

### 2. Бронирование ресторана (Booking)

#### Участники:
- клиент
- представитель

#### Этапы:
- выбор требуемого ресторана и предподчительного времени для посещения
- создание запроса на бронирование
- рассмотрение запроса о бронировании
- подтверждение запроса
- отмена запроса и предложение альтернатив для бронирования
- получение статуса заявки 

### 3. Посещение ресторана и формирование отзыва (Visiting and Feedback)

#### Участники:
- клиент
- представитель

#### Этапы:
- посещение ресторана
- подтверждение посещения клиентом
- подтверждение посещения представителем 
- формированием отчета представителем
- формирование отзыва клиентом
- формирование ответа представителя


### 4. Поиск/выбор подходящего ресторана (Search restaurant)

#### Участники:
- клиент

#### Этапы:
- выбор кухни
- выбор уровня стоимости 
- выбор непосредственно ресторана
